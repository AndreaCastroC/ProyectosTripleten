## Etapa 4. Prueba de hipótesis <a id='hypothesis'></a>

El siguiente apartado tiene como objetivo probar la siguiente hipótesis **La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare cambia los sábados lluviosos**, para poder responder a lo anterior, se realizará una prueba t para muestras independientes.

```python
#Filtrado para los viajes en sábados con lluvia
sabados = clima[clima["start_ts"].dt.weekday == 5]
sabados_lluviosos = sabados[sabados["weather_conditions"].str.contains("Bad", case=False, na=False)]
sabados_no_lluviosos = sabados[sabados["weather_conditions"].str.contains("Good", case=False, na=False)]

print(sabados_lluviosos.head())
print(sabados_no_lluviosos.head())
```

```python
#Duración de los viajes en sábados lluviosos y no lluviosos
duracion_lluviosos = sabados_lluviosos["duration_seconds"]
duracion_no_lluviosos = sabados_no_lluviosos["duration_seconds"]
```

Para poder realizar la prueba t, es necesario realizar primero la prueba de Levene. Para ello se plantean las siguientes hipótesis:

La **H0** es que la varianza entre la duración promedio de los viajes desde Loop hasta el Aeropuerto Internacional O'Hare los sábados lluviosos y la varianza entre la duración promedio de los viajes desde Loop hasta el Aeropuerto Internacional O'Hare los sábados no lluviosos son iguales. 

La **H1** es que la varianza entre la duración promedio de los viajes desde Loop hasta el Aeropuerto Internacional O'Hare los sábados lluviosos y la varianza entre la duración promedio de los viajes desde Loop hasta el Aeropuerto Internacional O'Hare los sábados no lluviosos no son iguales

```python
#Prueba de Levene
stat, p_value_levene = levene(duracion_lluviosos, duracion_no_lluviosos)
print("Estadístico de Levene:", stat)
print("Valor p:", p_value_levene)
```

Siendo el p-valor mayor a 0.05, no podemos rechazar la H0. Por tanto, no podemos afirmar que exista igualdad de varianzas entre la duración promedio de los viajes desde Loop hasta el Aeropuerto Internacional O'Hare los sábados lluviosos y la duración promedio de los viajes desde Loop hasta el Aeropuerto Internacional O'Hare los sábados no lluviosos, ni tampoco lo contario.

A continuación se establecen las hipótesis para la prueba T:
La **H0** es que no hay diferencia entre la duración promedio de los viajes desde Loop hasta el Aeropuerto Internacional O'Hare los sábados lluviosos y la duración promedio de los viajes desde Loop hasta el Aeropuerto Internacional O'Hare los sábados no lluviosos. 

La **H1** es que sí hay diferencia entre la duración promedio de los viajes desde Loop hasta el Aeropuerto Internacional O'Hare los sábados lluviosos y la duración promedio de los viajes desde Loop hasta el Aeropuerto Internacional O'Hare los sábados no lluviosos. 

```python
#Prueba T
equal_var = p_value_levene > 0.05
stat_ttest, p_value_ttest = ttest_ind(duracion_lluviosos, duracion_no_lluviosos, equal_var=equal_var)
print(f"Prueba t: p-value = {p_value_ttest}")

#Establecemos el valor de alpha
alpha = 0.05

#Conclusión de la prueba t
if p_value_ttest < alpha:
    print("Rechazamos la hipótesis nula.\n")
else:
    print("No podemos rechazar la hipótesis nula.\n")
```

```python
#Gráfico complementario: barras de error

mean_lluviosos = sabados_lluviosos["duration_seconds"].mean()
mean_no_lluviosos = sabados_no_lluviosos["duration_seconds"].mean()

sem_lluviosos = sabados_lluviosos["duration_seconds"].sem()
sem_no_lluviosos = sabados_no_lluviosos["duration_seconds"].sem()
```

```python
data = {
    "Clima": ["Sábados lluviosos", "Sábados no lluviosos"],
    "Promedio de duración, en segundos": [mean_lluviosos, mean_no_lluviosos],
    "SEM": [sem_lluviosos, sem_no_lluviosos]
}

grafico = pd.DataFrame(data)
```

```python
#Gráfico de barras de error

plt.figure(figsize=(8, 6))

plt.errorbar(x=range(len(grafico)), y=grafico["Promedio de duración, en segundos"], 
             yerr=grafico["SEM"], fmt="none", c="black", capsize=5)

plt.xticks(range(len(grafico)), grafico['Clima'])

plt.title("Duración promedio de viajes, en segundos, los sábados")
plt.xlabel("Clima")
plt.ylabel("Duración promedio (segundos)")

plt.tight_layout()
plt.show()
```

En base al resultado obtenido podemos afirmar que se rechaza la hipótesis nula y que, por tanto, **existe una diferencia estadísticamente significativa** entre la duración promedio de los viajes desde Loop hasta el Aeropuerto Internacional O'Hare los sábados lluviosos y la duración promedio de los viajes desde Loop hasta el Aeropuerto Internacional O'Hare los sábados no lluviosos. En efecto, el gráfico de barras de error permite afirmar que durante los sábados lluviosos los viajes suelen tener una mayor duración promedio, quizá ello se deba a que los conductores toman mayores medidas de seguridad dadas las condiciones climáticas.
